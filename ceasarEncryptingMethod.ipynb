```python
import ceasarArt

alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm',
            'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

# def shiftLetterToRight(position, shiftValue):
#     if position + shiftValue >= len(alphabet):
#         return position + shiftValue - len(alphabet)
#     return position + shiftValue

# def shiftLetterToLeft(position, shiftValue):
#     return position - shiftValue  

def shiftLetter(position, shiftValue, direction):
    if direction == "encode":
        if position + shiftValue >= len(alphabet):
            return position + shiftValue - len(alphabet)
        return position + shiftValue
    elif direction == "decode":
        return position - shiftValue

# def encrypt(text, shift):
#     encryptedMessage = ""
#     for l in text:
#         shiftedIndex = shiftLetterToRight(alphabet.index(l), shift)
#         encryptedMessage += alphabet[shiftedIndex]
#     print(f"The encoded text is {encryptedMessage}")

# def decode(text, shift):
#     decryptedMessage = ""
#     for l in text:
#         shiftedIndex = shiftLetterToLeft(alphabet.index(l),shift)
#         decryptedMessage += alphabet[shiftedIndex]
#     print(f"The encoded text is {decryptedMessage}")

def ceasar(text, shift, cipherDirection):
    cipherMessage = ""
    for l in text:
        
        if l in alphabet:
            shiftedIndex = shiftLetter(alphabet.index(l), shift, cipherDirection)
            letter = alphabet[shiftedIndex]
        else:
            letter = l
        
        cipherMessage += letter
    
    print(f"The {cipherDirection}d message is: {cipherMessage}")

print(ceasarArt.logo)


while True:
    
    print("--------------------------------------------------------------")
    
    direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n")
    text = input("Type your message:\n").lower()
    shift = int(input("Type the shift number:\n"))
    
    #Hugely large shift value
    shift = shift % len(alphabet)
    
    ceasar(text, shift, direction)
    
    again = input("\nStart again?: 'Y' or 'N'\n").lower()
    
    if again != "y":
        break
        
        
```
