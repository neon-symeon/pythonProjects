```python
import pandas as pd
import random as rn
from hangman_art import stages, logo
from hangman_words import word_list

df = pd.read_csv('slowa.txt', names=['Col'], header=None)
df.head()
word_list = df["Col"].values.tolist()

game_on = True

while game_on:
    print(logo)
    print('\nZaczynamy grę. Zgadnij albo... zgiń :)')
    print(f'\nWczytuję bazę {len(word_list)} słów, więc nie będzie łatwo.\nSkoncentruj się :)')

    #wprowadzanie słowa
    while True:
        try:
            word_max_len = 0
            while word_max_len < 2:
                word_max_len = int(input('\nJak długie słowo maksymlanie jesteś gotów zgadywac?  '))  
        except:
            print('** BŁĄD ** . podaj maksymalną LICZBĘ liter w wyrazie')
        else:
            break

    wybrane_slowo_len = 100

    while word_max_len < wybrane_slowo_len:
        wybrane_slowo = rn.choice(word_list).lower()
        wybrane_slowo_len = len(wybrane_slowo)
    
    print(f'Wylosowane słowo ma: {wybrane_slowo_len} liter. Powodzenia :)')

    #tworzy pustą ramkę do zgadywania
    display = []
    for _ in range(wybrane_slowo_len):
        display += "_"

    guess_tested = []

    lives_left = len(stages) - 1
    przelot = 0

    while lives_left>0:# and przelot<12:

        przelot +=1

        #wyświetla wypełnione (tyle o ile) słowo
        print(f"{' '.join(display)}")

        #wprowadzanie litery do testowania
        while True:
            guess = input('podaj literę: ').lower()
            if len(guess)>1:
                print(f'\n wprowadż ponownie ale JEDNĄ literę, teraz wprowadziłeś {len(guess)}')
            elif guess in guess_tested:
                print(f'\n hej, już testowałaś/łeś tę literę. wprowadź inną niż {guess_tested}')
            else:
                guess_tested.append(guess)
                break
    
        #sprawdzanie czy litera jest w słowie
        for position in range(wybrane_slowo_len):
            letter = wybrane_slowo[position]
            if letter == guess:
                display[position] = letter
                print('\nWspaniale. Trafione.')

        #działania gdy jednak pudło
        if guess not in wybrane_slowo:
            lives_left += -1
            if lives_left == 1:
                print('Hej. To już ostatnia próba. Skup się i walcz o życie :)')
            else:
                print(f'\nSorki. Pudło. Pozostało prób {lives_left}')

        if not "_" in display:
            print(f'wygrana!! odgadłeś słowo :) G R A T U L A C J E !!! w {przelot} próbach. WOW :)')
            break
        
        #jeśli przegrana
        if lives_left == 0:
            print(f'Przegrana. Neodgadnione słowo to {wybrane_slowo.upper()}.\n')
        
        print(stages[lives_left])

    print(f'jeśli trafiło Ci się wyjątkowo trune słowo,\njego znaczenie sprawdzisz na:\nhttps://sjp.pl/{wybrane_slowo}')
    czy_gramy_dalej = input('\następna gra? t / n').lower()
    
    if czy_gramy_dalej == 'n':
        game_on = False
        print('KOŃCZYMY. DZIĘKI ZA GRĘ. DO NASTĘPNEGO RAZU! MIŁEGO DNIA.')
    
```
