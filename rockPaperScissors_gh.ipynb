```python
# work paper scissors associacion, rules, tournaments and more: https://wrpsa.com/

# good luck!

ROCK = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

PAPER = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

SCISSORS = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

TROPHY = '''

 .-=========-.
  \-=======-/
 _|   .=.   |_
((|  {{1}}  |))
 \|   /|\   |/
  \__ '`' __/
    _`) (`_
  _/_______\_
 /___________\

'''

#imports
import random as rn

#results of the competition between the user and Python :)
result = [0,0]

#all possible hands
hand_set = [ROCK, PAPER, SCISSORS]

# ---- the beginning
while True:
    #coputer choice
    cpu_hand = rn.choice(hand_set)
    
    print('----------------------------------------------------------')
    print('----------------------------------------------------------')
    
    #user choice
    while True:
        players_hand = input('Your hand?\nr - for rock\np  -for paper or\ns - for scissors\n   ').lower()
        if players_hand == 'r':
            usr_hand = ROCK
            break
        elif players_hand == 'p':
            usr_hand = PAPER
            break
        elif players_hand == 's':
            usr_hand = SCISSORS
            break
        else:
            print('write "p", "r", or "s", other letters generate error')

    #hand's comparison
    print('----------------------------------------------------------')
    
    if usr_hand == cpu_hand:
        result[0] += 1
        result[1] += 1
        print(
            f"it's a DRAW. Both your and computer's hands are the same\n {usr_hand}")

    elif usr_hand == ROCK:
        if cpu_hand == SCISSORS:
            result[0] += 1
            print(
                f"*** You WIN!!! *** \n\n Your hand is a ROCK \n {usr_hand} \n and computer's hand are SCISSORS {cpu_hand}"
            )
        else:
            result[1] += 1
            print(
                f"*** You LOOSE!!! *** \n\n Your hand is a ROCK \n {usr_hand} \n and computer's hand is a PAPER {cpu_hand}"
            )

    elif usr_hand == SCISSORS:
        if cpu_hand == PAPER:
            result[0] += 1
            print(
                f"*** You WIN!!! *** \n\n Your hand are SCISSORS \n {usr_hand} \n and computer's hand is a sheet of PAPER {cpu_hand}"
            )
        else:
            result[1] += 1
            print(
                f"*** You LOOSE!!! *** \n\n Your hand are SCISSORS \n {usr_hand} \n and computer's hand is a ROCK {cpu_hand}"
            )

    elif usr_hand == PAPER:
        if cpu_hand == ROCK:
            result[0] += 1
            print(
                f"*** You WIN!!! *** \n\n Your hand is a sheet of PAPER \n {usr_hand} \n and computer's hand is a ROCK {cpu_hand}"
            )
        else:
            result[1] += 1
            print(
                f"*** You LOOSE!!! *** \n\n Your hand is a sheet of PAPER \n {usr_hand} \n and computer's hand is a SCISSORS {cpu_hand}"
            )
    
    print('*******************')
    print(f'Your Score:     {result[0]}')
    print(f'Computer Score: {result[1]}')
    print('*******************')

    #play again?
    play_again = input('\nPlay again? y, n\n\n').lower()

    if play_again == 'n':
        if result[0]>result[1]:
            print("\nCongratulations. You WON the GAME\n")
            print()
            print(TROPHY)
            print()
        elif result[0]<result[1]:
            print("\nDon't worry. You'd be better next time.\n")
        else:
            print("\nThe Draw means you are in the flow. try again and win :)\n ")

        #print('\nGame Over. Thank you. See U next time :)')
        break
```
